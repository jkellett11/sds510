```python
#importing libraries
from selenium import webdriver
from selenium.webdriver.common.by import By #find_element_by_ID was depreciated
from time import sleep
import pandas as pd
import csv
import config
```

```python
#writing an empty csv to store later
with open('output.csv', 'w') as f:
    writer=csv.writer(f)
    writer.writerow(['post'])
```

```python
#telling Selenium to open Chrome
driver = webdriver.Chrome()
```

```python
#Navigating to the social media site
driver.get('https://mastodon.social/auth/sign_in')
```

```python
#inputing the username in the username box
log_form = driver.find_element(By.ID,'user_email').send_keys(config.UNAME)
sleep(20)
```

```python
#inputing the password in the password box
pass_form = driver.find_element(By.ID, 'user_password').send_keys(config.PASSWORD)
sleep(15)
```

```python
#clicking the login button
pass_button = driver.find_element(By.CLASS_NAME, 'btn').click()
```

```python
#defining the keyword I am searching for
keyword = 'harkins theater'
```

```python

search = driver.find_element(By.CLASS_NAME, "search__input").send_keys(keyword + '\n')
```

```python
postbutton = driver.find_element(By.XPATH, "//div[@class='account__section-headline']/button[4]/span[text()='Posts']").click()
sleep(10)
```

```python
#getting relevant public posts
postdata = driver.find_elements(By.XPATH, "//div[@class='status status-public']/div[@class='status__content status__content--with-action']/div[@class='status__content__text status__content__text--visible translate']")
```

```python
#appending the text data of each post to a list
postlist = []
for eachpost in postdata:
    post = eachpost.text
    postlist.append({'post':post})
```

```python
#creating a dataframe
df = pd.DataFrame(postlist)
```

```python
#displaying the dataframe
df
```

```python
#exporting the dataframe to a csv
df.to_csv('output.csv', index=False)
```

```python
#quitting the driver
driver.quit()
```
